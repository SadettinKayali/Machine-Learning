```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv("DecisionTrees_titanic.csv")
dataset.head()
```

```python
X = dataset.drop("Survived", axis=1)
y = dataset.loc[:,"Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
decisiontree = DecisionTreeClassifier()
randomforest = RandomForestClassifier()
```

```python
decisiontree.fit(X_train, y_train)
randomforest.fit(X_train, y_train)
```

```python
decisiontree_pred = decisiontree.predict(X_test)
randomforest_pred = randomforest.predict(X_test)
```

```python
print(classification_report(y_test, decisiontree_pred))
```

```python
print(classification_report(y_test, randomforest_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))
plot_tree(decisiontree,
          max_depth=2,
          feature_names=X.columns,
          filled=True,
          impurity=False,
          rounded=True,
          precision=1)

plt.show()
```

Decision tree algorithms are a series of if-else statements that can be used to predict a result based on data. Simply put, it randomly asks questions, checks if the dataset is divided correctly, and changes the questions accordingly.

The advantages of decision tree algorithms are that there is no need to normalize or scale the data, and missing values don’t significantly affect its performance. On the other hand, decision trees usually take longer to train, and they tend to overfit complicated data.

Decision trees can be used for solving classification as well as regression problems.

For decision trees, we can perform an operation called “pruning”. Pruning is basically removing unnecessary features based on the information gain, in order to get a less complex decision tree.

Random forest uses many trees instead of using only one. Therefore, it is more complex than decision trees and is less affected by noise. Since we have many trees in random forest and each tree focuses on different parts of the dataset, this prevents overfitting.
